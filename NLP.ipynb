```python
# NLP/NLTK
# Created on Nov 9 2019 by Athulya Gopishetty
   
# for OS interface (to get/change directory)
import os  
# import the NLTK
import nltk

# display and load the NLTK source site
nltk.download()

# display and set working/data directory
os.getcwd()
os.chdir('C:\\Users\\rahul\\Desktop\\AIT580\\DATA')
os.getcwd()

# open the text file and display
f = open('BikeInjury.txt')
text_file = f.read()
# display the 'text_file' contents
print(text_file)
f.close()

# convert 'text_file' contents to lower case and display
text_file = text_file.lower()
print(text_file)

from nltk.tokenize import sent_tokenize, word_tokenize

# extracting and displaying sentences and words from 'text_file'
sentences_text = sent_tokenize(text_file)
print(sentences_text)
words_text = word_tokenize(text_file)
print(words_text)

# Count and graph the word frequency  
freq_dist1 = nltk.FreqDist(words_text)
print(freq_dist1)
print(freq_dist1.most_common(25))
freq_dist1.plot(25)

# storing all strings in alphabets# in 'bike_injury'
# eliminating non-alphabets in strings
bike_injury = [w for w in words_text if w.isalpha()]

from nltk.corpus import stopwords

# Get rid of "stop words" and store the data in 'filtered_words'
text_stopwords = set(stopwords.words('english'))
filtered_words = []

for w in bike_injury: 
    if w not in text_stopwords:
        filtered_words.append(w)
        
from nltk.stem import WordNetLemmatizer

# lemmatize the words and display
word_lemmatizer = WordNetLemmatizer()
filtered_words = [word_lemmatizer.lemmatize(word) for word in filtered_words]
print(filtered_words)

filtered_text_words = nltk.Text(filtered_words)

# search words similar to injuries if any
filtered_text_words.similar('injuries')

from nltk.stem.porter import PorterStemmer

# reducing derived words using stemming process
port = PorterStemmer()
stem = [port.stem(word) for word in filtered_text_words]
filtered_text_words = nltk.Text(stem)

from nltk.book import FreqDist

# display and plot the 25 most common injuries 
freq_dist2 = FreqDist(stem)
print(freq_dist2.most_common(25))
freq_dist2.plot(25)

# display several injury description sentences using the sent_tokenize()
words = ['trauma','elbow']
sentences = sent_tokenize(text_file)
for sentence in sentences:
    if(all(map(lambda word: word in sentence, words))):
        print(sentence)


# In[ ]:

```
